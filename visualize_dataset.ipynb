# Visualization

```python
import sys
sys.path.append('src')

import matplotlib.pyplot as plt

from dataset.cifar10 import CIFAR10

dataset = CIFAR10(train=False, transform=False)
plt.figure(figsize=(10, 1))
for i in range(8):
    img, label = dataset[i]
    print(img)  # PIL Image
    plt.subplot(1, 8, i+1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(dataset.id_to_class[label])
```

```python
import sys
sys.path.append('src')

import matplotlib.pyplot as plt

from dataset.imagenet import ImageNet2012

dataset = ImageNet2012(train=False, transform=False)
plt.figure(figsize=(8, 2))
for i in range(3):
    img, label = dataset[100+i]
    print(img)  # PIL Image
    plt.subplot(1, 3, i+1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(dataset.id_to_class[label][0])
```
